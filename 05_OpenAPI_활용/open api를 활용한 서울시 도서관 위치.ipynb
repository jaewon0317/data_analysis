```python
import requests
import pandas as pd
```

```python
url = 'http://openapi.seoul.go.kr:8088/6976495561636c6f38324e68585967/json/SeoulPublicLibraryInfo/1/200/'
response = requests.get(url)
print(response.content)
```

```python
data = response.json()
data
```

```python
temp = []
for d in data['SeoulPublicLibraryInfo']['row']:
    temp.append(d)
lib_seoul = pd.DataFrame(temp)
lib_seoul
```

```python
lib_seoul.info()
```

```python
import folium
map = folium.Map(location=[37.4965,126.9572],   
                 zoom_start=11,                 
                 tiles='cartodbpositron')       
```

```python
map = folium.Map(location=[37.4965,126.9572],   
                 zoom_start=11,                 
                 tiles='cartodbpositron') 
for n, row in lib_seoul.iterrows():
    mk = folium.Marker([row['XCNTS'],row['YDNTS']],popup = folium.Popup(row['LBRRY_NAME'],max_width=200))
    mk.add_to(map)
map
```
